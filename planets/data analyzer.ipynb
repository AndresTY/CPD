```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
file = os.listdir('results/')
print(file)
```

```python
files =['planeta-500000.csv', 'planeta-1000000.csv','planeta-5000000.csv', 'planeta-10000000.csv', 'planeta-50000000.csv']
```

```python
x=[]
cy=[]
py=[]
for path in files:
    df = pd.read_csv('results/'+path)
    x.append(df['steps'].mean())
    cy.append(df['cython'].mean())
    py.append(df['python'].mean())
```

```python
plt.plot(x, py, label='Python')
plt.plot(x, cy, label='Cython')
plt.xlabel('steps')
plt.ylabel('time (s)')
plt.title('python vs cython')
plt.legend()
plt.grid()
plt.show()
```

```python
x_t=[]
ty=[]
for path in files:
    df = pd.read_csv('results/'+path)
    x_t.append(str(df['steps'].mean()))
```

```python
dif = np.subtract(py,cy)
```

```python
x_p = np.arange(len(x_t))
plt.bar(x_p+ 0.2,py,0.4, label='Python')
plt.bar(x_p - 0.2,cy,0.4, label='Cython')
plt.plot(x_p, dif, label='diference between python and cython values',color='r')
plt.xticks(x_p, x_t)
plt.title('Comparison and difference in performance')
plt.xlabel('tests')
plt.ylabel('time(s)')
plt.grid()
plt.legend()
plt.show()
```
